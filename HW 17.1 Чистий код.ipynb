````python
import numpy as np
import pandas as pd

np.random.seed(0)

products = ["apple", "banana", "cherry", "dates", "elderberry"]
categories = ["fruit", "fruit", "fruit", "fruit", "fruit"]

N_ROWS = 10**6

df = pd.DataFrame({
    "product_name": np.random.choice(products, N_ROWS),
    "product_category": np.random.choice(categories, N_ROWS),
    "product_price": np.random.uniform(1, 10, N_ROWS),
    "transaction_date": pd.date_range(
        start="1/1/2000",
        periods=N_ROWS,
        freq="h"
    )
})

# Скільки використовується пам'яті до оптимізації
memory_before = df.memory_usage(deep=True).sum()

# Перетворення текстових стовпців у category
df["product_name"] = df["product_name"].astype("category")
df["product_category"] = df["product_category"].astype("category")

# Використання пам'яті після оптимізації
memory_after = df.memory_usage(deep=True).sum()

# Обчислення зменшення використання пам'яті
memory_reduction = memory_before / memory_after

print(f"Memory before optimization: {memory_before / 1024**2:.2f} MB")
print(f"Memory after optimization: {memory_after / 1024**2:.2f} MB")
print(f"New DataFrame uses memory in {memory_reduction:.2f} times less")
````

````python
def calculate_variance(numbers):
    """Обчислює дисперсію списку чисел."""

    total = 0

    for number in numbers:
        total += number

    mean = total / len(numbers)

    squared_diff_sum = 0

    for number in numbers:
        squared_diff_sum += (number - mean) ** 2

    variance = squared_diff_sum / len(numbers)

    return variance


input_list = [75, 80, 85, 90, 95]

result = calculate_variance(input_list)

print(result)
````
